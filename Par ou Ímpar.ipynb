```python
import random
c =0

while True:
    jogador = int(input('Digite um número'))
    ip = str(input('Você escolhe Par (P) ou Ímpar (I)')).upper().strip()
    máquina = random.randint(0, 10)
    soma = jogador + máquina
    resto = soma % 2

    if resto == 0 and ip == 'P':
        c += 1
        print(f'Você escolheu {jogador} e o computador {máquina}. A Soma dos dois números é {soma}. Deu PAR')
        print('Você VENCEU!')
        

    elif resto != 0 and ip == 'I':
        c += 1
        print(f'Você escolheu {jogador} e o computador {máquina}. A Soma dos dois números é {soma}. Deu ÍMPAR')
        print('Você VENCEU!')

    else:
        if resto == 0 and ip != 'P':
            print(f'Você escolheu {jogador} e o computador {máquina}. A Soma dos dois números é {soma}. Você escolheu ÍMPAR')
            print('Você PERDEU')
            print(f'Você ganhou {c} vezes.')

        elif resto != 0 and ip != 'I':
            print(f'Você escolheu {jogador} e o computador {máquina}. A Soma dos dois números é {soma}. Você escolheu PAR')
            print('Você PERDEU')
            print(f'Você ganhou {c} vezes.')
        break
```
